# Duplicate Cleaning Notebook (ID-Level Integrity)

This notebook focuses on resolving **duplicate records based on unique ID constraints**.

### Objectives:
- Detect duplicate IDs
- Investigate conflicting records
- Resolve duplicates using deterministic rules
- Ensure ID uniqueness constraint
- Export cleaned dataset

## Step 1: Load Dataset

```python
import pandas as pd
import numpy as np

df = pd.read_csv('../../data/raw/dog_survey.csv')
df.columns = df.columns.str.strip().str.lower()

df.head()
```

## Step 2: Detect Duplicate IDs

```python
# Check duplicate IDs
id_counts = df['id'].value_counts()
duplicate_ids = id_counts[id_counts > 1]

print("Total duplicate IDs:", len(duplicate_ids))
print("\nSample duplicate IDs:")
print(duplicate_ids.head(10))

# Extract full duplicate records
df_dup = df[df['id'].isin(duplicate_ids.index)].sort_values('id')

df_dup
```

## Step 3: Inspect Conflicts Within Duplicate IDs

```python
# Show conflicting records per ID
for i in duplicate_ids.index[:5]:
    print("\n========================")
    print(f"ID: {i}")
    print(df[df['id'] == i])
```

## Step 4: Resolve Duplicate IDs

```python
# Strategy: keep first occurrence (deterministic)
# Alternative could be: keep last or most complete row

before_rows = len(df)

df = df.drop_duplicates(subset=['id'], keep='first')

after_rows = len(df)

print("Rows before:", before_rows)
print("Rows after:", after_rows)
print("Duplicates removed:", before_rows - after_rows)
```

## Step 5: Validate ID Uniqueness

```python
assert df['id'].is_unique, "ID still contains duplicates!"

print("ID uniqueness confirmed.")
```

## Step 6: Export Cleaned Dataset

```python
output_path = '../../data/processed/duplicates_dog_survey_cleaned.csv'
df.to_csv(output_path, index=False)

print(f"Saved: {output_path}")
```
